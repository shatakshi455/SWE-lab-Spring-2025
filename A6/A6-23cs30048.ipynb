elements in the array.
Hint: For the data, use the .csv file “Book1.csv”

3. For the sorted Numpy 1-D array as obtained in Q.2, reverse the array and print.

```python
len = sorted.shape[0]
rev = np.zeros((len))
for i in range(len):
    rev[i] = sorted[len-1-i]
print(rev)
```

6. Write a program to convert a color image (say a.PNG) into a grescale image. Let the greysacle
image stored in the Numpy 2-D array be X. Display the grayscale iamge on the screen.
Hint: Greyscale value of a pixel is the mean of three RGB values of that pixel

```python
import numpy as np
gray_img = np.zeros((image_rgb.shape[0], image_rgb.shape[1]))
for i in range(image_rgb.shape[0]):
    for j in range(image_rgb.shape[1]):
        gray_img[i,j] = (image_rgb[i,j,0]//3 + image_rgb[i,j,1]//3 + image_rgb[i,j,2]//3)
plt.imshow(gray_img, cmap='gray')
plt.axis('off')
plt.show()
X = gray_img
```

11. Using the grayscale image stored in X, transform it into the binarized image with thresholds:
[50, 70, 100, 150]. Let the binarized images are stored in Z50, Z70, Z100, and Z150,
respectively.
Hint: Binarizing is thresholding each pixel value, i.e., if pixel>threshold, then 1 else 0.

```python
Z50 = X
Z70 = X
Z100 = X
Z150 = X
for i in range(X.shape[0]):
    for j in range(X.shape[1]):
        
        if(X[i][j]>=50):
            Z50[i][j] = 1
        else:
            Z50[i][j] = 0
            
        if(X[i][j]>=100):
            Z100[i][j] = 1
        else:
            Z100[i][j] = 0
            
        if(X[i][j]>=70):
            Z70[i][j] = 1
        else:
            Z70[i][j] = 0

        if(X[i][j]>=150):
            Z150[i][j] = 1
        else:
            Z150[i][j] = 0
print("Binarized!")
```

12. Consider the color image stored in a.png. Create a filter of [[-1,-1,-1][0,0,0][1,1,1]], and
multiply this filter to each pixel value in the image. Display the image after filtering.

```python
fil = np.array([[-1, -1, -1], [0, 0, 0], [1, 1, 1]])
img = cv2.imread('a.png')
image_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
new_img = image_rgb
for i in range(image_rgb.shape[0]):
    for j in range(image_rgb.shape[1]):
        new_img[i][j] = np.matmul(image_rgb[i][j], fil)
plt.imshow(new_img)
plt.axis('off')
plt.show()
```
